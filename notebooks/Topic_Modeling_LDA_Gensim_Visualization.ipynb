# LDA Coherence Search 1 Results
- Show results of first search through topic numbers to find the optimal coherence score, then examine the resulting topics for that topic number and coherence score. 
- With this search we are looking at topic values from 2 to 32, at 3 topic intervals.

## Import Data

```python
optimal_model = pickle.load(open('../data/pickles/lda/lda_gen_coherence_optimal_model_1.pkl', 'rb'))
corpus = pickle.load(open('../data/pickles/lda/lda_gen_bigrams_corpus.pkl', 'rb'))
id2word = pickle.load(open('../data/pickles/lda/lda_gen_bigrams_id2word.pkl', 'rb'))
start, limit, step = pickle.load(open('../data/pickles/lda/lda_gen_coherence_topic_search_values_1.pkl', 'rb'))
coherence_values = pickle.load(open('../data/pickles/lda/lda_gen_coherence_model_values_1.pkl', 'rb'))
```

## Show Coherence

```python
print("Top Coherence Score:", max(coherence_values))
print("Optimal Topic Number:", str(coherence_values.index(max(coherence_values)) * step + start))
```

```python
# Show graph of coherence values for each n-gram
x = range(start, limit, step)
plt.figure(figsize=(10,7))
plt.plot(x, coherence_values)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.show()
```

## Show Topics

```python
# Print topics of optimial model
pprint(optimal_model.print_topics(num_topics=-1)) 
```

- There are a couple of somewhat intelligible topics here. 
    - Topic 0 seems to be making a negative association betwee Disney and franchising, probably speaking to criticism of Disney monetizing and commodifying the franchise.
    - Topic 7 seems to be talking about socio-political issues with 'pc' 'popular_culture' and 'heavy-handed', indicating they believe that there was a deliberate attempt to make the film very politically correct.
    

## Visualize Topic Separations in pyLDAvis

```python
# Display model results plot
pyLDAvis.enable_notebook()
vis_2 = pyLDAvis.gensim.prepare(optimal_model, corpus, id2word)
vis_2
```

- Some solid separation in the top 3 topics, but significant overlap in the others, despite the high level of coherence.
